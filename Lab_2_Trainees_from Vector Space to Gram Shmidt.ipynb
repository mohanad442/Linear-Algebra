## Determinant
### Find Determinant of:

$$\begin{bmatrix} 3&8 \\ 4&6 \end{bmatrix} , \begin{bmatrix} 4&6 \\ 3&8 \end{bmatrix}$$
$$\begin{bmatrix} 6&1&1 \\ 4&-2&5 \\ 2&8&7\end{bmatrix}$$
$$\begin{bmatrix} 3&2&0&1 \\ 4&0&1&2 \\ 3&0&2&1 \\ 9&2&3&1\end{bmatrix}$$


```python
A = np.array([[3,8],
              [4,6]])

B = np.array([[4,6], 
              [3,8]])

C = np.array([[6, 1, 1],
              [4, -2, 5],
              [2, 8, 7]])

D = np.array([[3, 2, 0, 1], 
              [4, 0, 1, 2], 
              [3, 0, 2, 1],
              [9, 2, 3, 1]])
```

```python
print("1. ", np.linalg.det(A))
print("2. ", np.linalg.det(B))
print("3. ", np.linalg.det(C))
print("4. ", np.linalg.det(D))
```
